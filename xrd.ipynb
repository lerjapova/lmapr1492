## Téléchargement du fichier cif

Comme pour les tâches précédentes, je vais commencer par télécharger le fichier cif qui correspond au matériau qui m'a été attribué. Dans mon cas, c'est le InP.

```python
from pymatgen.ext.matproj import MPRester
from pymatgen.analysis.diffraction.xrd import XRDCalculator
import matplotlib.pyplot as plt


with MPRester("m2OqTvaFpA9jexGPK") as m:

    #structure du matériau via son ID
    structure = m.get_structure_by_material_id("mp-966800")

structure.to(filename="InP.cif") #cette ligne de code permet de télécharger sous fichier cif le matériau
```

## Pics du diffractogramme

La tâche de cette semaine consiste à déterminer les 3 premiers pics du diffractogramme correspondant à mon matérai (en indiquant les indices hkl associés) pour un rayonnement X incident Cu Kα de longueur d’onde λ=1.54060 Å. Pour cela, nous allons utiliser le module analysis.diffraction.xrd de pymatgen. Ensuite, avec matplotlib, je vais représenter ces trois pics sur un graphe. 

#### Premier pic

```python
xrd = XRDCalculator(1.54060)
pattern = xrd.get_pattern(structure,two_theta_range=(0, 90))

print("Angle 2 theta en degrés : ",pattern.x[0])
print("Intensité relative en %: ", pattern.y[0])
print("Indices hkl : ",pattern.hkls[0])
print("Distance d : ",pattern.d_hkls[0])
```

#### Deuxième pic

```python
print("Angle 2 theta en degrés : ",pattern.x[1])
print("Intensité relative en %: ", pattern.y[1])
print("Indices hkl : ",pattern.hkls[1])
print("Distance d : ",pattern.d_hkls[1])
```

#### Troisième pic

```python
print("Angle 2 theta en degrés : ",pattern.x[2])
print("Intensité relative en %: ", pattern.y[2])
print("Indices hkl : ",pattern.hkls[2])
print("Distance d : ",pattern.d_hkls[2])
```

#### Graphe

```python
figure = xrd.show_plot(structure,two_theta_range=(23,28),annotate_peaks=True,ax=None,with_labels=True)
```
